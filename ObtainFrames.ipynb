```python
import cv2
import pickle
import numpy as np
%pylab inline
```

```python
vidcap = cv2.VideoCapture('data/test5.mp4')
success,image = vidcap.read()
count = 0
success = True
while success:
    image = cv2.cvtColor(image,cv2.COLOR_BGR2GRAY)
    image = cv2.resize(image,(0,0),fx=0.20,fy=0.20)
    outpath='data/vid5/'
    cv2.imwrite(outpath+"%d.jpg" % count, image)     # save frame as JPEG file
    success,image = vidcap.read()
    #print ('Read a new frame: ', success)
    count += 1
print (count)
#rename -e 's/\d+/sprintf("%02d",$&)/e' -- *.jpg
```
